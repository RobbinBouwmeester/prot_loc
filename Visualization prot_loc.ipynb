```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/feature_matrix.csv",index_col=0)
```

```python
prosite_out = open("data/prosite_filtered.csv","w")
prosite = open("data/prosite.csv")
```

```python
exclude_prosite = list(df.columns[df.sum() == 0])
for line in prosite:
    if line.split("\t")[0] in exclude_prosite: continue
    prosite_out.write(line)
prosite_out.close()
```

```python
f1 = "percentile_25_0.05_polarity"
f2 = "percentile_25_0.2_buriedresidues"
f2 = "seq_length"

plt.scatter(df[df["class"] == 0][f1],
            df[df["class"] == 0][f2],
            s=1.25,
            label="Cytoplasm (class 0)")

plt.scatter(df[df["class"] == 1][f1],
            df[df["class"] == 1][f2],
            s=1.25,
            label="Transmembrane (class 1)")

plt.xlabel("5th percentile polarity 25 AA rolling")
plt.ylabel("20th percentile buriedresidues 25 AA rolling")
plt.legend()
plt.savefig("figs/percentile_feats.png")
```

```python
features = ["length",
            "percentile_25_0.05_polarity",
            "percentile_25_0.2_buriedresidues",
            "D",
            "hphob_doolittle|7",
            "C",
            "polarity|5",
            "polarity|2",
            "polarity|4",
            "percentile_16_0.4_recognitionfactors",
            "percentile_2_0.6_molwt",
            "percentile_25_0.2_polarity"]

for f in features:
    sns.kdeplot(df[df["class"] == 1][f],label="Transmembrane (class 1)")
    sns.kdeplot(df[df["class"] == 0][f],label="Cytoplasm (class 0)")
    plt.title(f)
    plt.show()
```
